```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report
```

```python

# ✅ 1. تحميل بيانات Train و Validation
X_train = pd.read_csv("C:/LINCS_DATA/project/features/X_train.csv")
X_val = pd.read_csv("C:/LINCS_DATA/project/features/X_val.csv")
y_train = pd.read_csv("C:/LINCS_DATA/project/features/y_train.csv")["label"]
y_val = pd.read_csv("C:/LINCS_DATA/project/features/y_val.csv")["label"]
```

```python
# ✅ 2. تدريب Random Forest باستخدام class_weight فقط
clf = RandomForestClassifier(class_weight='balanced', random_state=42)
clf.fit(X_train, y_train)

```

```python

# ✅ 3. التقييم على Validation
y_val_pred = clf.predict(X_val)
y_val_prob = clf.predict_proba(X_val)[:, 1]
```

```python
# ✅ 4. عرض النتائج
print("=== Evaluation with class_weight='balanced' only ===")
print("Accuracy:", accuracy_score(y_val, y_val_pred))
print("F1 Score:", f1_score(y_val, y_val_pred))
print("ROC-AUC:", roc_auc_score(y_val, y_val_prob))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
print("\nClassification Report:\n", classification_report(y_val, y_val_pred))
```
